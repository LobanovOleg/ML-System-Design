4. Нормализация

```python
xmax = df['Open'].max()
xmin = df['Open'].min()

def norm(x):
    x = (x - xmin) / (xmax - xmin)
    return x

df['Open'] = df['Open'].apply(norm)
print(df)
```

5. Что такое SMA (Simple Moving Average) с окном 12,15,30.

Простая скользящая средняя является наиболее распространенным типом используемой средней. В SMA мы выполняем суммирование последних точек данных и делим их на период времени (12, 15 и 30 являются такими периодами). Чем выше значение скользящей ширины, тем больше данные сглаживаются, но огромное значение может привести к снижению неточности.

6. Построить график с ценой закрытия и посчитаным SMA 12,15,30

```python
df['SMA'] = df.Close.sub(df.Close.shift())
df['SMA'].plot(figsize=(20, 12))
df['SMA'].rolling(window=12).mean().plot()
df['SMA'].rolling(window=15).mean().plot()
df['SMA'].rolling(window=30).mean().plot()
```
